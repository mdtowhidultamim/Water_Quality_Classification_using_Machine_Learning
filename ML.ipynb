#5: Data Cleaning

##5.1 Create a Copy of the Original Dataset

```python
# ============================================================
# 5.1 Create a Copy of the Original Dataset
# ============================================================
# A copy of the original dataframe is created before cleaning.
# This is a good academic and practical practice because it preserves
# the original dataset and allows comparison before and after cleaning.

df_cleaned = df.copy()

print("A copy of the original dataset has been created for cleaning.")
print("Original dataset shape:", df.shape)
print("Cleaning dataset shape:", df_cleaned.shape)
```

##5.2 Replace Invalid Values

```python
# ============================================================
# 5.2 Replace Invalid Values
# ============================================================
# The dataset contains invalid values represented as '#NUM!'.
# These values are not valid numerical values and must be replaced
# with NaN so that they can be handled properly as missing values.

invalid_value = "#NUM!"

# Count invalid values before replacement.
invalid_before = (df_cleaned == invalid_value).sum().sum()

print("Total invalid '#NUM!' values before replacement:", invalid_before)

# Replace '#NUM!' with NaN.
df_cleaned.replace(invalid_value, np.nan, inplace=True)

# Count invalid values after replacement.
invalid_after = (df_cleaned == invalid_value).sum().sum()

print("Total invalid '#NUM!' values after replacement:", invalid_after)
```

##5.3 Convert Columns to Numeric

```python
# ============================================================
# 5.3 Convert Columns to Numeric
# ============================================================
# All columns should be numeric for machine learning model training.
# Some columns such as 'ammonia' and 'is_safe' may originally appear
# as object type because of invalid '#NUM!' values.
#
# errors='coerce' ensures that any remaining non-numeric values are
# converted into NaN.

for column in df_cleaned.columns:
    df_cleaned[column] = pd.to_numeric(df_cleaned[column], errors="coerce")

print("All columns have been converted to numeric format.")
```

##5.4 Check Data Types After Conversion

```python
# ============================================================
# 5.4 Check Data Types After Conversion
# ============================================================
# This step confirms whether all columns have been successfully
# converted into numerical data types.

print("Data types after numeric conversion:")
print(df_cleaned.dtypes)

# Identify any remaining object type columns.
remaining_object_columns = df_cleaned.select_dtypes(include="object").columns.tolist()

print("\nRemaining object type columns:", remaining_object_columns)

if len(remaining_object_columns) == 0:
    print("All columns are now numeric.")
else:
    print("Some object columns still remain and require further checking.")
```

##5.5 Check Missing Values After Conversion

```python
# ============================================================
# 5.5 Check Missing Values After Conversion
# ============================================================
# After replacing '#NUM!' and converting columns to numeric format,
# missing values are checked again.
#
# Missing values may appear because invalid values were converted to NaN.

missing_values = df_cleaned.isnull().sum()

print("Missing values per column:")
print(missing_values)

print("\nTotal missing values in the dataset:", df_cleaned.isnull().sum().sum())

# Display only columns that contain missing values.
columns_with_missing = missing_values[missing_values > 0]

print("\nColumns with missing values:")
print(columns_with_missing)
```

##5.6 Display Rows Containing Missing Values

```python
# ============================================================
# 5.6 Display Rows Containing Missing Values
# ============================================================
# This step displays the rows that contain missing values.
# Since only a very small number of rows are expected to contain
# missing values, these rows can be safely removed.

rows_with_missing = df_cleaned[df_cleaned.isnull().any(axis=1)]

print("Number of rows containing missing values:", rows_with_missing.shape[0])

rows_with_missing
```

##5.7 Handle Missing Values

```python
# ============================================================
# 5.7 Handle Missing Values
# ============================================================
# Rows containing missing values are removed.
#
# Justification:
# The number of invalid/missing rows is very small compared to the
# total dataset size. Therefore, removing these rows will not
# significantly affect the dataset but will ensure clean input data
# for machine learning models.

shape_before_missing_removal = df_cleaned.shape

df_cleaned.dropna(inplace=True)

shape_after_missing_removal = df_cleaned.shape

print("Shape before removing missing values:", shape_before_missing_removal)
print("Shape after removing missing values:", shape_after_missing_removal)
print("Number of rows removed:", shape_before_missing_removal[0] - shape_after_missing_removal[0])
```

##5.8 Reset Dataset Index

```python
# ============================================================
# 5.8 Reset Dataset Index
# ============================================================
# After removing rows, the dataframe index is reset to maintain
# a clean and continuous index sequence.

df_cleaned.reset_index(drop=True, inplace=True)

print("Dataset index has been reset successfully.")
print("Current dataset shape:", df_cleaned.shape)
```

#7: Class Distribution Analysis

```python
# ============================================================
# 7.1 Count Target Classes
# ============================================================
# This step counts the number of samples in each target class.
#
# Target variable:
# is_safe
#
# Class meanings:
# 0 = Not Safe
# 1 = Safe

target_column = "is_safe"

# Count the number of samples for each class.
class_counts = df_cleaned[target_column].value_counts().sort_index()

print("Class Counts for Target Variable:")
print("---------------------------------")
print(class_counts)

print("\nClass Meaning:")
print("0 = Not Safe")
print("1 = Safe")
```

##7.2 Calculate Class Percentage

###8.3.2 Create Outlier Summary Using IQR Method

```python
# ============================================================
# 8.3.2 Create Outlier Summary Using IQR Method
# ============================================================
# The Interquartile Range (IQR) method is used to estimate possible outliers.
#
# IQR = Q3 - Q1
# Lower bound = Q1 - 1.5 * IQR
# Upper bound = Q3 + 1.5 * IQR
#
# Values outside these bounds are considered possible outliers.

outlier_summary = []

for column in feature_columns:
    Q1 = df_cleaned[column].quantile(0.25)
    Q3 = df_cleaned[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_count = ((df_cleaned[column] < lower_bound) | (df_cleaned[column] > upper_bound)).sum()
    outlier_percentage = (outlier_count / df_cleaned.shape[0]) * 100

    outlier_summary.append({
        "Feature": column,
        "Q1": Q1,
        "Q3": Q3,
        "IQR": IQR,
        "Lower Bound": lower_bound,
        "Upper Bound": upper_bound,
        "Outlier Count": outlier_count,
        "Outlier Percentage (%)": round(outlier_percentage, 2)
    })

outlier_summary_table = pd.DataFrame(outlier_summary)

outlier_summary_table = outlier_summary_table.sort_values(
    by="Outlier Count",
    ascending=False
).reset_index(drop=True)

print("Outlier Summary Using IQR Method:")
outlier_summary_table
```

#12: Outlier Detection and Handling

##12.1 Detect Outliers Using Boxplots

```python
# ============================================================
# 12.1 Detect Outliers Using Boxplots
# ============================================================
# Boxplots are used to visualize the spread of feature values and to detect potential outliers.
# Outliers are defined as values that fall outside the whiskers of the boxplot (1.5 * IQR above Q3 or below Q1).

# Create boxplots for each numerical feature to detect outliers
plt.figure(figsize=(20, 12))
df_cleaned[feature_columns].boxplot(rot=90)

plt.title("Boxplot Analysis for Outlier Detection", fontsize=16)
plt.xlabel("Features", fontsize=12)
plt.ylabel("Feature Values", fontsize=12)
plt.tight_layout()
plt.show()
```

##12.2 Detect Outliers Using IQR

```python
# ============================================================
# 12.2 Detect Outliers Using IQR
# ============================================================
# The Interquartile Range (IQR) method is used to detect outliers by calculating the range between Q1 (25th percentile)
# and Q3 (75th percentile), then defining outliers as values outside of the range: Q1 - 1.5*IQR to Q3 + 1.5*IQR.

outlier_summary = []

for column in feature_columns:
    Q1 = df_cleaned[column].quantile(0.25)
    Q3 = df_cleaned[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_count = ((df_cleaned[column] < lower_bound) | (df_cleaned[column] > upper_bound)).sum()
    outlier_percentage = (outlier_count / df_cleaned.shape[0]) * 100

    outlier_summary.append({
        "Feature": column,
        "Q1": Q1,
        "Q3": Q3,
        "IQR": IQR,
        "Lower Bound": lower_bound,
        "Upper Bound": upper_bound,
        "Outlier Count": outlier_count,
        "Outlier Percentage (%)": round(outlier_percentage, 2)
    })

outlier_summary_table = pd.DataFrame(outlier_summary)

# Sorting by the outlier count for better visibility
outlier_summary_table = outlier_summary_table.sort_values(
    by="Outlier Count",
    ascending=False
).reset_index(drop=True)

print("Outlier Summary Using IQR Method:")
outlier_summary_table
```

##12.3 Decide Outlier Treatment

```python
# ============================================================
# 12.3 Decide Outlier Treatment
# ============================================================
# Based on the outlier summary table, we decide how to treat the outliers.
# In this case, we will not remove outliers automatically, as high pollutant values
# may be valid, indicating unsafe water conditions.

print("Outlier Treatment Decision:")
print("----------------------------")

# We choose not to remove outliers automatically, but rather assess their impact.
print("Decision: We will not remove all outliers automatically.")
print("Justification: High pollutant values may represent real unsafe water conditions.")
```

##12.4 Justify Decision

```python
# ============================================================
# 12.4 Justify Decision
# ============================================================
# The decision not to remove all outliers automatically is justified,
# as extreme pollutant levels may actually reflect unsafe water samples.
# Therefore, these values are important and should not be discarded unless there is clear evidence of data errors.

print("Justification for Outlier Treatment Decision:")
print("-------------------------------------------")

print(
    "High pollutant values in water quality data, such as high levels of mercury, arsenic, or nitrates, "
    "may indicate unsafe water conditions. These extreme values should not automatically be treated as errors."
)
print(
    "Removing these outliers without careful consideration could lead to incorrect conclusions about water safety."
)
```

##15.8 Optional KNN Tuning

```python
# ============================================================
# 15.8 Optional KNN Tuning
# ============================================================
# In this step, we will test different values of k (the number of neighbors)
# and select the best k based on the F1-score.
#
# The goal is to choose the best k that balances precision and recall, improving overall model performance.

from sklearn.model_selection import cross_val_score
import numpy as np

# Test different values of k (from 1 to 20) and evaluate using cross-validation
k_values = range(1, 21)
f1_scores = []

for k in k_values:
    knn_model = KNeighborsClassifier(n_neighbors=k)
    f1_scores.append(np.mean(cross_val_score(knn_model, X_train_knn, y_train_knn, cv=5, scoring='f1')))

# Find the best k based on the maximum F1-score
best_k = k_values[np.argmax(f1_scores)]

print(f"Best k value based on F1-score: {best_k}")
```

##16.8 Optional Decision Tree Tuning

```python
# ============================================================
# 16.8 Optional Decision Tree Tuning
# ============================================================
# In this step, we will tune the Decision Tree model by adjusting its hyperparameters, such as:
# - Maximum depth of the tree
# - Minimum samples required to split a node
# - Minimum samples required to be at a leaf node
# - The criterion used for the split (Gini impurity or entropy)

from sklearn.model_selection import GridSearchCV

# Define the hyperparameters to tune
param_grid = {
    "max_depth": [3, 5, 7, 10, None],  # Maximum depth of the tree
    "min_samples_split": [2, 5, 10],  # Minimum samples required to split a node
    "min_samples_leaf": [1, 2, 4],    # Minimum samples required to be at a leaf node
    "criterion": ["gini", "entropy"] # Criterion for the split
}

# Initialize GridSearchCV to search over hyperparameters
grid_search = GridSearchCV(estimator=dt_model, param_grid=param_grid, cv=5, scoring="f1", verbose=1)

# Fit GridSearchCV to the training data
grid_search.fit(X_train_dt, y_train_dt)

# Best hyperparameters from GridSearchCV
best_params = grid_search.best_params_

print("Best hyperparameters found through GridSearchCV:")
print(best_params)
```
